```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 공부한 시간에 따른 합격 유무 (1: 합격, 0: 불합격)

x = np.array([2,4,6,8,10,12,14])  # 공부한 시간
y = np.array([0,0,0,1,1,1,1])     # 합격 유무
```

```python
# 모델 실행

model = Sequential()
model.add(Dense(1, input_dim=1, activation='sigmoid'))

# Sequential() 모델 특징
# 입력층이 따로 없고, 반드시 첫번째 Dense layer에 input_dim 이 나옴
# model.add(Dense(출력, 입력, 활성화 함수))

model.compile(optimizer='sgd', loss='binary_crossentropy')
# optimizer(최적화): sgd 경사하강법으로 미세조정 (fine tuning)
# loss(loss function, cost function 비용함수 >> y - y_pred 최소화)
# binary(이진의) cross entropy(교차 엔트로피)
```

```python
# 학습
model.fit(x,y, epochs=5000)
```

```python
# 모델 테스트 데이터 시각화

plt.scatter(x,y) # 실제 값(산점도)
plt.plot(x, model.predict(x), 'r') # 예측값
plt.show()
```

```python
# 임의의 학습시간 입력, 합격 예상 확률 예측

hour = 7
prediction = model.predict([hour])

print("%d 시간을 공부할 경우, 당신의 합격 예상 확률은 %.2f입니다"%(hour, prediction*100))
```
