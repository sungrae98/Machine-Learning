- 표준화(Standardization)

```python
import seaborn as sns

iris = sns.load_dataset('iris')
x = iris.iloc[:,:-1]
y = iris.iloc[:,-1]
```

```python
print(x) # 속성 정보
print(x.shape)
print()
print(y) # 종속 정보 target
print(y.shape)
```

```python
# 데이터 전처리 : scailing

from sklearn.preprocessing import scale

scale(x)
```

```python
x_scaled = scale(x)
```

```python
x_scaled[:5, :]
```

```python
x_scaled.mean(axis = 0)
# axis = 0 (열 기준, 서로 다른 행끼리)
# axis = 1 (행 기존, 서로 다른 열끼리)
# default(기본값) >> axis = 0
```

```python
# min-max scailing

from sklearn.preprocessing import minmax_scale

sc = minmax_scale(x)
sc[:5, :]
# 0 ~ 1 사이 값으로 스케일링
```

```python
x[:5]
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(x)
# 표준화(평균, 표준편차 계산)
# fit을 한다는 것은 영점 조절을 했다는 의미
```

```python
iris_scaled = sc.transform(x)
```

```python
iris_scaled[:5, :]
```

```python
# scale back(스케일 백)
# >> 원래 값을 알고 싶어요
# inverse_trainsform()

sc.inverse_transform(iris_scaled)[:5,:]
```
