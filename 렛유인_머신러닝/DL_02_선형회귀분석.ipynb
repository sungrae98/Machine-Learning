```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 공부시간과 성적의 관계

x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# 기울기 w, 절편 b

# 초기화
w = 0
b = 0

# 학습률(learning rate) (임의로 0.03)
lr = 0.03

# 몇 번 반복 횟수 정하기
epochs = 2001
```

```python
# 경사하강법

n = len(x)

# 경사하강법 시작

for i in range(epochs):
     y_pred = w * x + b
     error = y - y_pred

     # 편미분 구하기
     w_diff = (2/n) * sum(-x * (error)) # 오차함수를 w로 편미분한 값
     b_diff = (2/n) * sum(-(error))     # 오차함수를 b로 편미분한 값

     # 경사하강을 하면서 업데이트
     # 학습률 곱한 뒤 기존의 w, b 업데이트
     w  = w - lr * w_diff
     b = b - lr * b_diff

     if i % 100 == 0 :
        print('epoch=%.f, 기울기=%.4f, 절편=%.4f'%(i, w, b))
```

```python
# epoch=1500, 기울기=2.3000, 절편=79.0000
# y(성적) = 2.3 x(공부한 시간) + 79
```

```python
y_pred = w * x + b

plt.scatter(x,y)
plt.plot(x, y_pred, 'r') # r: red
plt.show()
```

다중 선형회귀

```python
x1 = np.array([2,4,6,8])      # 공부한 시간
x2 = np.array([0,4,2,3])      # 과외 받은 시간(횟수)
y = np.array([81,93,91,97])   # 실제 받은 성적
```

```python
# 데이터 분포 확인

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

```python
# 변수 설정

w1 = 0
w2 = 0
b = 0

# 학습률 설정(임의로 정함)
lr = 0.01

# 몇 번 반복할 지 설정
epochs = 40001
```

```python
# 경사하강법

n = len(x1)

for i in range(epochs):
    y_pred = w1 * x1 + w2 * x2 + b
    error = y - y_pred

    # 편미분
    w1_diff = (2/n) * sum(-x1 * (error))
    w2_diff = (2/n) * sum(-x2 * (error))
    b_diff = (2/n) * sum(-(error))

    # 학습률에 따라 가중치(w), y 절편(b) 업데이트
    w1 = w1 - lr * w1_diff
    w2 = w2 - lr * w2_diff
    b = b - lr * b_diff

    if i % 100 == 0 :
        print('epoch:%.f, 기울기 w1:%.4f, 기울기 w2:%.4f, 절편 b:%.4f'%(i, w1, w2, b))
```

```python
# 실제 점수와 예측점수 출력

print('실제점수: ', y)
print('예측점수: ', y_pred)
```

```python
# 텐서플로우 활용, 단순 선형회귀 모델

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2,4,6,8])       # 공부한 시간
y = np.array([81,93,91,97])   # 실제 성적
```

```python
model = Sequential()

model.add(Dense(1, input_dim = 1, activation ='linear'))
# Dense(출력 값, 입력 변수, 분석 방법 순) 설정

model.compile(optimizer='sgd', loss='mse')
# 오차 수정 하기 위해 경사하강법 사용(sgd) (최적화)
# 오차 정도를 판단하기 위한 평가 방법 >> mse(평균제곱오차) 사용

model.fit(x, y, epochs=2001)
# 학습 오차를 최소화는 과정으로 2000번 반복
```

```python
# 모델 테스트

hour = 7

prediction = model.predict([hour])

print('%.f 시간을 공부하면, 예상점수는 %0.2f 점입니다'%(hour, prediction))
```

다중선형회귀 분석

```python
x = np.array([[2,0],[4,4],[6,2],[8,3]]) #  [공부시간, 과외 횟수]
y = np.array([81,93,91,97])
```

```python
model = Sequential()

model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer ='sgd', loss='mse')

model.fit(x,y, epochs=2001)
```

```python
# 모델 테스트

hour = 7
private_class = 4

prediction = model.predict([[hour, private_class]])

print('%d 시간을 공부하고, %d시간 과외를 받으면, 예상점수는 %0.2f 점입니다.'%(hour,
                                                       private_class, prediction))
```
